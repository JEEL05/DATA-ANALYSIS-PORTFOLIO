```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("FMCG_2022_2024.csv")


df['date'] = pd.to_datetime(df['date'], errors='coerce')
df = df.dropna(subset=['date']).sort_values('date').reset_index(drop=True)
```

```python

df['month'] = df['date'].dt.to_period('M')
df['year'] = df['date'].dt.year
df['weekday'] = df['date'].dt.day_name()
```

```python

sku_sales = df.groupby('sku')['units_sold'].sum().sort_values(ascending=False)
print("Top SKUs:\n", sku_sales.head())

top_skus = sku_sales.head(5).index
monthly_trend = (df[df['sku'].isin(top_skus)]
                 .groupby(['month','sku'])['units_sold']
                 .sum()
                 .unstack(fill_value=0))
monthly_trend.plot(title="Monthly Units Sold Trend (Top 5 SKUs)")
plt.xticks(rotation=45)
plt.show()

```

```python
promo_impact = df.groupby('promotion_flag')['units_sold'].mean()
sns.barplot(x=promo_impact.index, y=promo_impact.values)
plt.title("Average Units Sold (Promo vs No Promo)")
plt.show()
```

```python
df['stockout'] = (df['stock_available'] == 0).astype(int)
stockout_counts = df.groupby('sku')['stockout'].sum().sort_values(ascending=False)
print("SKUs with most stockouts:\n", stockout_counts.head())
```

```python

df = df.sort_values(['sku', 'date']).reset_index(drop=True)

df['units_sold'] = pd.to_numeric(df['units_sold'], errors='coerce')


df['units_sold_lag_1'] = df.groupby('sku')['units_sold'].shift(1)

df['units_sold_ma_7'] = (
    df.groupby('sku')['units_sold']
      .shift(1)                                 
      .rolling(window=7, min_periods=1)        
      .mean()
      .reset_index(level=0, drop=True)
)

```

```python
some_sku = df['sku'].iloc[0]  
df_sku = df[df['sku'] == some_sku][['date','sku','units_sold','units_sold_lag_1','units_sold_ma_7']]

print(df_sku.head(15))
```

```python
df['units_sold_ma_3'] = (
    df.groupby('sku')['units_sold']
      .shift(1)
      .rolling(window=3, min_periods=1)
      .mean()
      .reset_index(level=0, drop=True)
)
```

```python
ax = df[df['sku']==some_sku].set_index('date')[['units_sold','units_sold_ma_7']].plot(title=f"{some_sku} - units_sold vs MA7")
ax.figure.autofmt_xdate()
```
